```python
import numpy as np
import matplotlib.pyplot as plt

def splineQuad(x,y,f_deriv,nb_point_int):
    n= len(x)
    #h = np.zeros(n-1)
    z=np.zeros(n)
    
    #intialisation de z[0] et calcul des autre valeur de zi
    
    z[0] = f_deriv(x[0])   # S'(x0)
    
    for i in range(n-1):
        
        z[i+1]=(2*(y[i+1]-y[i])/(x[i+1]-x[i]))-z[i]
        


    #initialisation  et calcul des coefficient de chaque spline
    a = np.zeros(n-1)
    b = np.zeros(n-1)
    c = np.zeros(n-1)
    
    for i in range(n-1):
    
        a[i]=(z[i+1] - z[i]) / (2*(x[i+1]-x[i]))
    
        b[i]=z[i]
    
        c[i]=y[i]
                

    
    # Affichage des fonctions Si(x)
    print("Les fonctions Si(x) pour chaque intervalle :")
    for i in range(n - 1):
        print(f"Sur [{x[i]}, {x[i+1]}] : Si(x) = {a[i]:.3f}*(x - ({x[i]}))² + {b[i]:.3f}*(x - {x[i]}) + {c[i]:.3f}")

    # Tracé graphique
    plt.figure(figsize=(8, 6))
    for i in range(n - 1):
        xi = np.linspace(x[i], x[i + 1], nb_point_int)
        Si = a[i] * (xi - x[i])**2 + b[i] * (xi - x[i]) + c[i]
        plt.plot(xi, Si, label=f"S{i}(x) sur [{x[i]}, {x[i+1]}]")


    # Tracer les points d'origine
    plt.scatter(x, y, color='red', label='Points donnés')
    plt.title("Représentation graphique de la spline quadratique")
    plt.xlabel("x")
    plt.ylabel("S(x)")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# Définition de la fonction
def f(x):
    return x**4

# Dérivée de f(x)
def f_deriv(x):
    return 4*x**3

# Liste des 5 points
x_vals = [-2,-1, 0, 1, 2]

# Calcul des images
for x in x_vals:
    print(f"f({x}) = {f(x)}")
    y_vals = [f(x) for x in x_vals]

splineQuad(x_vals,y_vals,f_deriv, 10000)

```
